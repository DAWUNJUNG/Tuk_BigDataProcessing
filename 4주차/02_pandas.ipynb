## 데이터 프레임 만들기
- 딕셔너리를 만든 후 이를 데이터프레임으로 바꾸는 방법
- csv 파일을 읽어 만드는 방법
- 액셀 파일을 읽어 만드는 방법
- 배열, 리스트, 튜플로부터 만드는 방법

```python
# 딕셔너리에서 만드는 방법
import pandas as pd

x = {'city': ['서울', '부산', '대구', '대전', '광주'],
        'population': [990, 350, 250, 154, 150],
        'temp': [13, 16, 14, 13, 15]}

data = pd.DataFrame(x)
data
```

## 샘플 추출
- 원하는 샘플(행)을 얻는 방법 : loc와 iloc 사용
- loc : 행(index)과 열(column name)에 이름 기반으로 접근
- iloc : 행(세로순번)과 열(가로순번)에 순서번호 기반으로 접근

## loc,iloc 모두 아래와 같은 형태로 접근 가능
- loc[r, c] --> 행, 열 지정
- loc[r1:r2, c] or loc[r, c1:c2]  --> (행범위 지정, 열지정) 혹은 (행지정, 열범위 지정)
- loc[r, :] or [:, c]  --> (행 지정, 열 전체) 혹은 (행 전체, 열 지정) 
- loc[r1:r2, c1:c2] --> (행범위 지정, 열범위 지정)


```python
# 인덱스 사용 (loc)
df.loc[['부산','울산']]
```

```python
# 행의 위치 순서 번호 사용
df.iloc[1:4]
```
